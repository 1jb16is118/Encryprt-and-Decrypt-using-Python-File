```python
import sys

################  data_import()   function  #####################
def data_import(fname):
    f=open(fname,'r')
    array=[]                # array to store data from txt file
    for i in f:
        array.extend(a for a in i.split())
    f.close()
    return array

##################################################################

##############   plain() function   #########################
def plain():
    f=open('plain.txt','r')
    data=f.read()
    f.close()
    return data
###############################################################
###############   my_key() function ###########################
def my_key(student_name,sid,age):
    sum1=0
   # print(sid)
   # print(len(sid))
    for i in sid:
        sum1=sum1+int(i)
        
    remainder=sum1%int(age)
    if remainder==0: 
        right_shift=sid[len(sid)-1]+sid[:len(sid)-1]
        key=right_shift+age+str(len(student_name))
    elif remainder<=9:
        left_shift=sid[remainder:len(sid)-1]+sid[:remainder]
        key=left_shift+age+str(len(student_name))
    elif remainder>9:
        remainder1=remainder-10
        
        left_shift=sid[remainder1:len(sid)-1]+sid[:remainder1]
        key=left_shift+age+str(len(student_name))
    
    
    f=open("mykey.txt",'w')
    f.write(key)
    f.close()
    return key

#################################################################

#########  rounding_average() function  ############################

def rounding_average(key):
    sum2=0
    count=0
    
    for i in key[:len(key)-1]:
        sum2=sum2+int(i)
        count=count+1
        average=sum2/count
        rounding=round(average)
    return rounding

#################################################################

############  caesar() function   #############################

def caesar(data,rounding):
    
    letter='abcdefghijklmnopqrstuvwxyz'
    x=[]
    data1=data.lower()
    
    #print(len(data1))
    for i in data1:
        index=letter.find(i)

        if index+rounding>26:
            x.append(letter[(index+rounding)%26])
        else:
            x.append(letter[index+rounding])

    result="".join(x)
    
    return result

#####################################################################

##########   ciphertxt() function  ###################################

def ciphertxt(key,cipher):
    count=1
    x=[]
    if len(key)==len(cipher):
        for i in range(len(key)):
            x.append(cipher[i])
            x.append(key[i])
        
        return "".join(x)
    elif len(key)>len(cipher):
        for i in range(len(cipher)):
            count+=1
            x.append(cipher[i])
            x.append(key[i])
        return "".join(x)+key[count:]
    elif len(key)<len(cipher):
        
        count1=0
        key1=''
        while(len(cipher)>=len(key1)):
            key1=key1+key
        
        for i in range(len(cipher)):
            count+=1
            x.append(cipher[i])
            x.append(key1[i])
        return "".join(x)+key1[count:]
    
#####################################################################

#########  decrypt  function  #######################################
def decrypt(cipher):
    string=[]
    for i in cipher:
        if i.isalpha():
            string.append(i)
    return "".join(string)

#####################################################################

##### caser1 function  ##############################################
def caser1(string,rounding):
    letter='abcdefghijklmnopqrstuvwxyz'
    x=[]
    
    for i in string:
        index=letter.find(i)

        if index+rounding>26:
            x.append(letter[(index-rounding)%26])
        else:
            x.append(letter[index-rounding])
    result="".join(x)
    
    return result
####################################################################
print("Menu")   # main
print("(1) encrypt\n(2) decrypt")    # choose option
value=input("enter any one above menu: ")
if value=="encrypt":
    ID=input("enter your id: ")       # Enter id For varification
    fname=ID+".txt"     # Create ID.txt file
    try:
        f=open(fname)
    
        import_data=data_import(fname) # calling data_import function
        student_name=import_data[0]
        sid=import_data[1]
        age=import_data[2]
        print('encryption\ndecryption\nexit')
        user=input("enter any one from above:")      # enter encryption or decryption
        if user=="encryption":
            data=plain()                      # calling Plain function
            key=my_key(student_name,sid,age)  # calling my_key function
            rounding=rounding_average(key)    # calling rounding_average function
            cipher=caesar(data,rounding)      # calling caeser function
            txt=ciphertxt(key,cipher)         # calling ciphertxt function
            g=open("cipher.txt",'w')          # writing encrypted txt in cipher.txt
            g.write(txt)
            g.close()
        elif user=='decryption':
            try:
                f=open("cipher.txt",'r')
                cipher=f.read()                   # Reading txt from cipher.txt
                f.close()
                f=open("mykey.txt",'r')
                key=f.read()                      # Reading key from mykey.txt
                f.close()

                string=decrypt(cipher)            # calling decrypt function

                rounding=rounding_average(key)

                original=caser1(string,rounding)

                f=open("original.txt",'w')
                f.write(original)
                f.close()
            except IOError:
                print("first do encryption")
        elif user=="exit":
            exit()
    
    except IOError:
        print("file does not exit")
        
elif value=='decrypt':
    try:
        f=open("cipher.txt",'r')
        cipher=f.read()
        f.close()
        f=open("mykey.txt",'r')
        key=f.read()
        f.close()

        string=decrypt(cipher)

        rounding=rounding_average(key)

        original=caser1(string,rounding)      # calling caser1 function

        f=open("original.txt",'w')
        f.write(original)
        f.close()
    except IOError:
        print("first do encryption")
    
```
